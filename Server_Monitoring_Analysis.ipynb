```python

import pandas as pd

df = pd.read_csv("Sample_Data_Ingestion.csv")
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Log_Timestamp"] = pd.to_datetime(df["Log_Timestamp"], dayfirst=True)
```

```python
df["Log_Timestamp"].head()
```

```python
df["Server_Status"] = df["CPU_Utilization (%)"].apply(
    lambda x: "Critical" if x > 85 else "Warning" if x > 70 else "Healthy"
)
```

```python
df.to_csv("cleaned_server_data.csv", index=False)
```
